```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Welcome to my first data visualization project in Python. In here, you will find graphs on select data in higher education across 7,592 colleges and universities in the United States. Data analysis utilizes enrollment count, average SAT scores, net cost of education, and salary after graduation. Data courtesy of Data.gov and IPEDS.

```python
higher_ed_stat = pd.read_csv('higher_education_data.csv',encoding ='latin1')
print(higher_ed_stat)
```

```python
list(higher_ed_stat)
```

```python
enrollment = higher_ed_stat.drop(['CITY','SAT_AVG','NPT4_PUB','NPT4_PRIV','MD_EARN_WNE_P10'], axis=1)
enrollment
```

```python
enrollment_no_null = enrollment.dropna(subset=['UGDS'])
enrollment_no_null
```

```python
enrollment_by_state = enrollment_no_null.groupby('STABBR').sum()
enrollment_by_state
```

# Enrollment by State

```python
enrollment_by_state['UGDS'].plot(kind = 'bar', figsize = (18,6), color = 'gray', fontsize = 15)
```

# As shown in the bar graph above, California is home to the largest student population, with numbers exceeding 2,000,000. This is followed by Texas, with numbers at approximately 1,300,000. Population-wise, California and Texas are also number 1 and 2 in ranking. Note that data from US territories were included.

```python
cost_and_income_public = higher_ed_stat.drop(['INSTNM','CITY','STABBR','SAT_AVG','UGDS','NPT4_PRIV'], axis=1)
cost_and_income_public
```

```python
cost_and_income_public_no_null = cost_and_income_public.dropna(subset=['NPT4_PUB','MD_EARN_WNE_P10'])
cost_and_income_public_no_null
```

```python
cost_and_income_public_clean = cost_and_income_public_no_null[cost_and_income_public_no_null['MD_EARN_WNE_P10']!='PrivacySuppressed']
cost_and_income_public_clean
```

# Does Cost Correlate with Future Income in Public Institutions?

```python
plt.scatter(x=cost_and_income_public_clean['NPT4_PUB'], y=cost_and_income_public_clean['MD_EARN_WNE_P10'], s=10)
plt.xlabel("Average Net Cost of Education")
plt.ylabel("Alumni Salary")
plt.title("Does Cost Correlate with Future Income in Public Institutions?")
plt.show()
```

# Well, the graph appears to be suggesting some correlation. But a linear regression analysis needs to be performed in order to confirm that. Statistical analysis will be covered in the next chapter of this data science prep bootcamp. Average Net Cost of Education substracts funding from government and other sources to the total cost of attendance. Salary was measured 10 years after completion of degree.

```python
SAT_and_income = higher_ed_stat.drop(['INSTNM','CITY','STABBR','UGDS','NPT4_PUB','NPT4_PRIV'], axis=1)
SAT_and_income
```

```python
SAT_and_income_no_null = SAT_and_income.dropna(subset=['SAT_AVG','MD_EARN_WNE_P10'])
SAT_and_income_no_null
```

```python
SAT_and_income_clean = SAT_and_income_no_null[SAT_and_income_no_null['MD_EARN_WNE_P10']!='PrivacySuppressed']
SAT_and_income_clean
```

```python
SAT_and_income_sort = SAT_and_income_clean.sort_values(by='SAT_AVG')
SAT_and_income_sort
```

# Does SAT Score Correlate with Future Income in Colleges and Universities?

```python
plt.plot(SAT_and_income_sort['SAT_AVG'],SAT_and_income_sort['MD_EARN_WNE_P10'],color='green')
plt.xlabel("Average SAT Score")
plt.ylabel("Alumni Salary")
plt.title("Does SAT Score Correlate with Future Income in Colleges and Universities?")
plt.show()
```

# The graph seems to suggest that this is the case. However, like in the previous plot, a linear regression analysis will confirm the hypothesis. The alumni salary was obtained 10 years after graduation.

```python
cost_of_college = higher_ed_stat.drop(['CITY','SAT_AVG','UGDS','MD_EARN_WNE_P10'], axis=1)
cost_of_college
```

```python
cost_of_college_public = cost_of_college.drop(['NPT4_PRIV'], axis=1)
cost_of_college_public
```

```python
cost_of_college_public_no_null = cost_of_college_public.dropna(subset=['NPT4_PUB'])
cost_of_college_public_no_null
```

```python
cost_of_college_public_by_state = cost_of_college_public.groupby('STABBR').mean()
cost_of_college_public_by_state
```

```python
cost_of_college_private = cost_of_college.drop(['NPT4_PUB'], axis=1)
cost_of_college_private
```

```python
cost_of_college_private_no_null = cost_of_college_private.dropna(subset=['NPT4_PRIV'])
cost_of_college_private_no_null
```

```python
cost_of_college_private_by_state = cost_of_college_private.groupby('STABBR').mean()
cost_of_college_private_by_state
```

# Net Cost of Attending College

```python
plt.figure(figsize=(40, 10))

plt.subplot(1, 2, 1)
cost_of_college_public_by_state['NPT4_PUB'].plot(kind = 'bar', color = 'red', fontsize = 20)
plt.ylim([0, 25000])
plt.ylabel('Average Net Cost of Attendance', fontsize = 30)
plt.xlabel('US State', fontsize = 30)
plt.title('Net Cost of Public Colleges by State', fontsize = 40)

plt.subplot(1, 2, 2)
cost_of_college_private_by_state['NPT4_PRIV'].plot(kind = 'bar', color = 'orange', fontsize = 20)
plt.ylim([0, 25000])
plt.ylabel('Average Net Cost of Attendance', fontsize = 30)
plt.xlabel('US State', fontsize = 30)
plt.title('Net Cost of Private Colleges by State', fontsize = 40)

plt.show()
```

# The bar graphs were plotted as a pair of subplots. The graph on the left shows average net cost of attendance among public institutions, averaged by state, while the graph on the right shows average net cost of attendance, also averaged by state. Clearly, students shell out more money, in general, by attending private institutions. In some States, the difference is larger than the net cost. The States of Rhode Island, Massachussetts, Washington DC, and Vermont, have the highest cost of attending private institutions.
